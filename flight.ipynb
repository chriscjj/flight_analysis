```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# 1. Flight Count by Airline
plt.figure(figsize=(10,5))
sns.countplot(y=fl_copy["Airline"], order=fl_copy["Airline"].value_counts().index, palette="viridis")
plt.xlabel("Number of Flights")
plt.ylabel("Airline")
plt.title("Flight Count by Airline")
plt.show()
```

```python
# 2. Average Flight Price by Airline
plt.figure(figsize=(10, 5))
fl_copy.groupby("Airline")["Price"].mean().sort_values().plot(kind="barh", color="skyblue")
plt.xlabel("Average Price")
plt.ylabel("Airline")
plt.title("Average Flight Price by Airline")
plt.show()
```

```python
# 3. Flight Duration Distribution
plt.figure(figsize=(10, 5))
sns.histplot(fl_copy["Duration"], bins=30, kde=True, color="blue")
plt.xlabel("Duration (minutes)")
plt.ylabel("Count")
plt.title("Flight Duration Distribution")
plt.xticks(rotation=180)
plt.show()
```
